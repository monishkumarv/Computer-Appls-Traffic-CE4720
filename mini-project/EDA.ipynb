```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset/all-crashes-2004-2018.csv', low_memory = False)
data = data[data['MAX_SEVERITY_LEVEL'] < 5]
```

```python
data['REAR_END_COLLISION'] = [1 if i == 1 else 0 for i in data['COLLISION_TYPE']]
data['HEAD_ON_COLLISION'] = [1 if i == 2 else 0 for i in data['COLLISION_TYPE']]
data['REAR2REAR_COLLISION'] = [1 if i == 3 else 0 for i in data['COLLISION_TYPE']]
data['ANGLE_COLLISION'] = [1 if i == 4 else 0 for i in data['COLLISION_TYPE']]
data['SIDESWEEP_COLLISION'] = [1 if (i == 5 or i == 6) else 0 for i in data['COLLISION_TYPE']]
```

```python
data['SMALL_VEH_COUNT'] = data["BICYCLE_COUNT"] + data["MOTORCYCLE_COUNT"]
data['MEDIUM_VEH_COUNT'] = data["AUTOMOBILE_COUNT"] + data["SUV_COUNT"]
data['LARGE_VEH_COUNT'] = data["VAN_COUNT"] + data["SMALL_TRUCK_COUNT"] + data["HEAVY_TRUCK_COUNT"] + data["BUS_COUNT"] + data["SCHOOL_BUS"]
```

```python
cols = ["MAX_SEVERITY_LEVEL",
        "AGGRESSIVE_DRIVING",
        
        "MC_DRINKING_DRIVER","DRINKING_DRIVER",
        
        "PED_COUNT","SMALL_VEH_COUNT","MEDIUM_VEH_COUNT","LARGE_VEH_COUNT",
        
        'HIT_BRIDGE','HIT_DEER','HIT_EMBANKMENT','HIT_FIXED_OBJECT','HIT_GDRAIL','HIT_GDRAIL_END',
        'HIT_POLE','HIT_TREE_SHRUB','HIT_BARRIER',
        
        "CROSS_MEDIAN","SV_RUN_OFF_RD",
        
        "DISTRACTED","SPEEDING","UNBELTED",        
        "FIRE_IN_VEHICLE","OVERTURNED",
       
        "REAR_END_COLLISION","HEAD_ON_COLLISION","REAR2REAR_COLLISION","ANGLE_COLLISION","SIDESWEEP_COLLISION"]
```

```python
data[cols].to_csv('filtered-data.csv', index = None)
```

```python
data['MAX_SEVERITY_LEVEL'] = data['MAX_SEVERITY_LEVEL'].replace(4,3)
data[cols].to_csv('filtered-data-combined.csv', index = None)
```
